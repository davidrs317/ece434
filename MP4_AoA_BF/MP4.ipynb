````python
def music_implemented():
    return False

def music_aoa(mic_data_folder, FS, MIC_OFFSETS):
    """MUSIC AoA algorithm. Write your code here.

    Args:
        mic_data_folder: name of folder (without a trailing slash) containing 
                         the mic datafiles `0.csv` through `7.csv` and `config.csv`.
        FS: microphone sampling frequency - 16kHz.
        MIC_OFFSETS: a list of tuples of each microphone's location relative to the center of its mic array. 
                     This list is calculated based on the diameter(0.09218m) and geometry of the microphone array.
                     For example, MIC_OFFSETS[1] is [0.09218*0.5, 0]. If the location of microphone array #i is
                     [x_i, y_i]. Then [x_i, y_i] + MIC_OFFSETS[j] yields the absolute location of mic#j of array#i.
                     This is provided for your convenience and you may choose to ignore.

    Returns:
        A list of incoming signal's polar angles for all mic arrays.

    """
    aoas = [0, 0, 0, 0, 0, 0, 0, 0] # Please comment out this line after you implement the function
    # Your return value should be the incoming signal's polar angle (between 0 and 2pi)
    return aoas

def shift_arr(arr, num, fill_value):
    result = np.empty_like(arr)
    if num > 0:
        result[:num] = fill_value
        result[num:] = arr[:-num]
    elif num < 0:
        result[num:] = fill_value
        result[:num] = arr[-num:]
    else:
        result[:] = arr
    return result

def calculate_correlation(signals):
    i = 0
    correlated = 0

    while i < len(signals):
        if i == 0:
            correlated = np.multiply((signals[i] - np.mean(signals[i])), (signals[i + 1] - np.mean(signals[i + 1])))
            i += 2
        else:
            correlated = np.multiply(correlated, (signals[i] - np.mean(signals[i])))
            i += 1
    
    return correlated

def delay_and_sum_aoa(mic_data_folder, FS, MIC_OFFSETS):
    """Delay and sum AoA algorithm. Write your code here.

    Args and output same as previous.
    
    """
    from scipy.fft import fft
    from scipy import signal as sps
    import math
    
    DIAMETER = 0.09218 # meters
    MIC_D = DIAMETER / 2 #hexagon is six equilateral triangles so distance between each mic in array is d/2
    UPSAMPLE = 10
    SOS = 343
    NUM_SENSORS = 6
    
    aoas = []
    
    for i in range(8):
        # read upsampled signals
        mic_df = pd.read_csv(f'{mic_data_folder}/{i}.csv', header=None)
        # antenna0 = signal.resample(mic_df[0].to_numpy(), len(mic_df[0])*UPSAMPLE)
        # antenna1 = signal.resample(mic_df[1].to_numpy(), len(mic_df[1])*UPSAMPLE)
        # antenna2 = signal.resample(mic_df[2].to_numpy(), len(mic_df[2])*UPSAMPLE)
        # antenna3 = signal.resample(mic_df[3].to_numpy(), len(mic_df[3])*UPSAMPLE)
        # antenna4 = signal.resample(mic_df[4].to_numpy(), len(mic_df[4])*UPSAMPLE)
        # antenna5 = signal.resample(mic_df[5].to_numpy(), len(mic_df[5])*UPSAMPLE)
        signals = []
        # read and upsample signals
        for column in mic_df:
            signals.append(sps.resample(mic_df[column].to_numpy(), len(mic_df[column]) * UPSAMPLE))
        
        c_thetas = []
        
        for theta in range(-180, 180, 1):
            # for i, signal in enumerate(signals):
            #     if i == 0:
            #         # calculate distance delay
            #         d_delay = -MIC_D * math.cos(math.radians(60 - theta))

            #         #then calculate time delay
            #         t_delay = d_delay / SOS

            #         # then calculate sample delay
            #         s_delay = t_delay * FS * UPSAMPLE

            #         # assign delay to its signal
            #         signal_to_delay[tuple(signal)] = s_delay
            #     elif i == 1:
            #         # calculate distance delay
            #         d_delay = -MIC_D * math.cos(math.radians(theta))

            #         # then calculate time delay
            #         t_delay = d_delay / SOS

            #         # then calculate sample delay
            #         s_delay = t_delay * FS * UPSAMPLE

            #         # assign delay to its signal
            #         signal_to_delay[tuple(signal)] = s_delay
            #     else:
            #         # calculate distance delay
            #         d_delay = -MIC_D * math.cos(math.radians((420 - (i * 60)) - theta))

            #         # then calculate sample delay
            #         t_delay = d_delay / SOS

            #         # then calculate sample delay
            #         s_delay = t_delay * FS * UPSAMPLE

            #         # assign delay to its signal
            #         signal_to_delay[tuple(signal)] = s_delay

            # moved to declaring things on their own instead of in loops for speed concerns

            # calculate distance delays where center is the reference
            d_delay0 = MIC_D * math.cos(math.radians(60 - theta))
            d_delay1 = MIC_D * math.cos(math.radians(theta))
            d_delay2 = MIC_D * math.cos(math.radians(300 - theta))
            d_delay3 = MIC_D * math.cos(math.radians(240 - theta))
            d_delay4 = MIC_D * math.cos(math.radians(180 - theta))
            d_delay5 = MIC_D * math.cos(math.radians(120 - theta))

            # calculate time delays
            t_delay0 = d_delay0 / SOS
            t_delay1 = d_delay1 / SOS
            t_delay2 = d_delay2 / SOS 
            t_delay3 = d_delay3 / SOS
            t_delay4 = d_delay4 / SOS
            t_delay5 = d_delay5 / SOS

            # calculate sample delays
            s_delay0 = t_delay0 * FS * UPSAMPLE; 
            s_delay1 = t_delay1 * FS * UPSAMPLE
            s_delay2 = t_delay2 * FS * UPSAMPLE
            s_delay3 = t_delay3 * FS * UPSAMPLE
            s_delay4 = t_delay4 * FS * UPSAMPLE
            s_delay5 = t_delay5 * FS * UPSAMPLE

            # shift signals
            shifted_signals = []
            shifted0 = shift_arr(signals[0], int(s_delay0), 0); shifted_signals.append(shifted0)
            shifted1 = shift_arr(signals[1], int(s_delay1), 0); shifted_signals.append(shifted1)
            shifted2 = shift_arr(signals[2], int(s_delay2), 0); shifted_signals.append(shifted2)
            shifted3 = shift_arr(signals[3], int(s_delay3), 0); shifted_signals.append(shifted3)
            shifted4 = shift_arr(signals[4], int(s_delay4), 0); shifted_signals.append(shifted4)
            shifted5 = shift_arr(signals[5], int(s_delay5), 0); shifted_signals.append(shifted5)

            correlation = calculate_correlation(shifted_signals)
            
            c_thetas.append(np.mean(correlation))
        # append angle in correct quadrant
        aoas.append(np.argmax(c_thetas) - 180)

    
    # aoas = [0, 0, 0, 0, 0, 0, 0, 0] # Please comment out this line after you implement the function
    # Your return value should be the incoming signal's polar angle (between 0 and 2pi)
    return np.radians(aoas)

def aoa_localization(aoas, mic_data_folder):
    """AoA localization algorithm. Write your code here.

    Args:
        aoas: output for aoa algorithm.
        mic_data_folder: same as previous.

    Returns:
        The user's location in this format (in metres): (L_x, L_y)

    """
    
    configuration = pd.read_csv(f'{mic_data_folder}/config.csv', header=None)
    x_vals = configuration[0].to_numpy()
    y_vals = configuration[1].to_numpy()
    scores = []
    coords = []
    
    x = min(x_vals)
    while x < max(x_vals):
        y = min(y_vals)
        while y < max(x_vals):
            # score is the mean of the difference between the current x,y
            # and the angles of arrival
            cur_scores = [abs(np.arctan2(y - configuration.iloc[i, 1], x - configuration.iloc[i, 0]) - aoas[i]) for i in range(len(aoas))]
            score_mean = np.mean(cur_scores)
            scores.append(score_mean)
            coords.append((x, y))
            y += 0.1
        x += 0.1
    
    ret_loc = np.argmin(scores)
    ret = coords[ret_loc]
    return ret
    
````
